```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salario_marcos = np.array([100, 200, 300, 400, 500, 150])
```

```python
salario_deyv = np.array([50, 300, 500, 150])
```

```python
# Add Label
plt.plot(salario_marcos, c='Black', ls ='--', marker ='s', ms =8, label='Sal치rios marcos')
plt.plot(salario_deyv, c='Red', ls ='--', marker ='s', ms =8, label='Sal치rios Deyv')
plt.show()
```

```python
# Add Legend
plt.plot(salario_marcos, c='Black', ls ='--', marker ='s', ms =8, label='Sal치rios marcos')
plt.plot(salario_deyv, c='Red', ls ='--', marker ='^', ms =8, label='Sal치rios Deyv')
plt.legend()
plt.show()
```

```python
# Add Legend upper left
plt.plot(salario_marcos, c='Black', ls ='--', marker ='s', ms =8, label='Sal치rios marcos')
plt.plot(salario_deyv, c='Red', ls ='--', marker ='^', ms =8, label='Sal치rios Deyv')
plt.legend(loc='upper left')
plt.show()
```

```python
salario_primo = [100, 200, 300, 500, 800, 150]
plt.plot(salario_marcos, c='Black', ls ='--', marker ='s', ms =8, label='Sal치rios marcos')
plt.plot(salario_deyv, c='Red', ls ='--', marker ='^', ms =8, label='Sal치rios Deyv')
plt.plot(salario_primo, c='Green', ls ='--', marker ='o', ms =8, label='Sal치rios Primo')
plt.legend(loc='upper left')
plt.show()
```
